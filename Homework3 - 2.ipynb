```python
# this allows plots to appear directly in the notebook
%matplotlib inline

import sklearn.datasets
import matplotlib.pyplot
import sklearn.cluster
```

```python
clusterer=sklearn.cluster.KMeans(n_clusters=3)
```

```python
ClusterCols = scorers1[['Tries','Career_games','T_percent']]
```

```python
cluster_suggestions=clusterer.fit_predict(ClusterCols)
cluster_suggestions
```

```python
import sklearn.metrics
```

```python
##### takes Hours!
sklearn.metrics.silhouette_score(ClusterCols, cluster_suggestions)
```

```python
dbscan=sklearn.cluster.DBSCAN(eps=0.5,min_samples=3)
```

```python
dbscan_suggestions=dbscan.fit_predict(ClusterCols)
dbscan_suggestions
```
